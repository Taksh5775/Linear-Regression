```python
sns.boxplot(data = df,x='area')
plt.show()
```

```python
Q1 = np.percentile(df['area'],25)
Q3 = np.percentile(df['area'],75)
print(Q1,Q3)
```

```python
IQR = Q3-Q1
l_fence = Q1 - (1.5*IQR)
u_fence = Q3 + (1.5*IQR)
print(l_fence,u_fence)
```
